```python
# | Feature                  | CSV                            | Excel                                         | JSON                             |
# | ------------------------ | ------------------------------ | --------------------------------------------- | -------------------------------- |
# | Full Form                | Comma-Separated Values         | Spreadsheet file (.xlsx, .xls)                | JavaScript Object Notation       |
# | Structure                | Plain text table               | Spreadsheet with sheets, formulas, formatting | Hierarchical key-value structure |
# | Human Readability        | Easy                           | Easy (in spreadsheet software)                | Moderate                         |
# | Supports Multiple Sheets | No                             | Yes                                           | No                               |
# | Supports Formatting      | No                             | Yes                                           | No                               |
# | Supports Formulas        | No                             | Yes                                           | No                               |
# | Supports Nested Data     | No                             | No (not naturally)                            | Yes                              |
# | File Size                | Usually small                  | Larger                                        | Moderate                         |
# | Common Use               | Data exchange, imports/exports | Reporting, analysis                           | APIs, web applications           |
#            csv                          excel                            json
# syntax:-   .csv                         .xlsx                            .json
```
