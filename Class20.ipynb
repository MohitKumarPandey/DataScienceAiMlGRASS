```python
x=[1,2,3,4]
arr=np.array(x)
y1=[10,13,23,12]
y2=[15,12,10,32]
y3=[40,30,20,10]
# calculations 
w=0.30
plt.figure(figsize=(6,4),dpi=90)
plt.bar(arr-w,y1,label='Boys',width=w)
plt.bar(arr,y2,label='Girls',width=w)
plt.bar(arr+w,y3,label='other',width=w)
plt.xlabel("Groups")
plt.ylabel("No. of students")
plt.title("No of group in each group")
plt.legend()
```

```python
# histogram chart
marks=[40,55,60,70,75,90,33,50]
plt.hist(marks,bins=2,color='g')# orientation='horizontal'
```

```python
# pie chart 
import matplotlib.pyplot as plt 
fruits=['Apple','Banana','orange','watermalon']
count=[40,30,15,70]
plt.pie(count,labels=fruits,autopct='%1.1f%%',colors=['red','yellow','orange','green'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = ['G1', 'G2', 'G3', 'G4']

boys = [10, 20, 15, 25]
girls = [5, 10, 12, 8]

# Plot
plt.bar(x, boys, label='Boys')
plt.bar(x, girls, bottom=boys, label='Girls')

plt.xlabel("Groups")
plt.ylabel("Students")
plt.title("Stacked Bar Chart")
plt.legend()

plt.show()
```

```python
boys = [10, 20, 15, 25]
girls = [5, 10, 12, 8]
others = [2, 3, 4, 5]
plt.bar(x, boys, label='Boys')
plt.bar(x, girls, bottom=boys, label='Girls')
plt.bar(x, others, bottom=np.array(boys)+np.array(girls), label='Others')
```
