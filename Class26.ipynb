```python
from sklearn.preprocessing import LabelEncoder 
import pandas as pd
import numpy as np 
```

```python
# handle missing values 
data={
    "color":['red','green','blue','orange','green','blue',np.nan]
}
df=pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
# label encoding 
df
le=LabelEncoder()
df['color_encoder']=le.fit_transform(df['color'])
df
```

```python
df['color_encoder1']=LabelEncoder().fit_transform(df['color'])
df
```

```python
import sklearn
df['color_encoder2']=sklearn.preprocessing.LabelEncoder().fit_transform(df['color'])
df
```

```python
# one hot encoding 
from sklearn.preprocessing import OneHotEncoder
if 'color_encoder' in df.keys():
    df.drop(df[['color_encoder','color_encoder1','color_encoder2']],axis=1,inplace=True)
```

```python
df
```

```python
oneHot=OneHotEncoder()
df1=oneHot.fit_transform(df[['color']])
df1.toarray()
```

```python
# by the help of pandas 
df2=pd.get_dummies(df,columns=['color'])
df2
```

```python
# combination of handle missing value,labelencoding ,one hot encoding 
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
data1={
    'age':[10,20,np.nan,26,30],
    'gender':['male','female','others','male',None],
    'name':['dheeraj','kavi','sapu','yash','hello']
}
df1=pd.DataFrame(data1)
# handle missing value 
df1['age']=df1['age'].fillna(df1['age'].mean())
df1.dropna(inplace=True)
df1
```

```python
# label encoding 
le=LabelEncoder()
df1['gender1']=le.fit_transform(df1['gender'])
df1
```

```python
# one hot encoding 
oe=OneHotEncoder()
df2=oe.fit_transform(df1[['gender']]).toarray()
df2
```
