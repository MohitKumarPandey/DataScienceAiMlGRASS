```python
# class 16
import pandas as pd 
d={
    "Name":["vishal","virat","vineet"],# 3 rows 
    "salary":[100,200,300]
    
}
df=pd.DataFrame(data=d) 
df["Holidays"]=df["salary"]/100
df["Decrement"]=[10,20,30]
df
```

```python
# delete column 
df.drop(["Decrement","Holidays"],axis=1,inplace=True)
df
# to delete from actual data we need to write inplace=True
```

```python
df
```

```python
# selection on rows 
df.loc[2,'Name']
df
```

```python
df.iloc[2,0]
```

```python
# get a single row data 
df.iloc[1]
print(df.loc[1])
```

```python
# get multiple rows 
print(df.iloc[0:3])
# get multi rows using loc 
print(df.loc[0:2])
```

```python
# sub data get using iloc 
# in iloc the last index in slicing is excluded 
df1=df.iloc[0:2,[0]]
df1
```

```python
# sub data get using loc 
# in loc the last index is included 
df2=df.loc[0:2,["Name"]]
df2
```

```python
# practice ques 
import pandas as pd

data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}

df3 = pd.DataFrame(data)

print(df3)
```

```python
print(df3['Name'])
print(df3[['Emp_ID','Salary']])
df3["Increment"]=df3["Salary"]+(df3["Salary"]*(30/100))
df3
# add rows 
df3.loc[6]=[107,'Jack','IT',60000,5,78000]
df3
# delete a single row 
df3.drop(7,inplace=True)
```

```python
df3
```

```python
# access using iloc 
df3.iloc[0]
df3.iloc[0:5]
df3.iloc[0,[1,2,3,4]]
```

```python
#  access using loc 
df3.loc[0]
df3.loc[[0],["Emp_ID","Name","Department","Salary"]]
# subdata 
```
