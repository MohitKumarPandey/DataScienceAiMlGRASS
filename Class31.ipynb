```python
# evaluation matrix 
import numpy as np
from sklearn.metrics import mean_absolute_error  # [MAE]
from sklearn.metrics import mean_squared_error    #[MSE]
from sklearn.metrics import r2_score             #[R2]
from sklearn.metrics import root_mean_squared_error  # [RSME]
```

```python
# actual values prediction 
y_true=np.array([50,60,75,90,100])

# predicted data
y_pred=np.array([52,58,78,88,105])

# mae(mean absolute error )
mae=mean_absolute_error(y_true,y_pred)
print(mae)
```

```python
# mse(mean square error)
mse=mean_squared_error(y_true,y_pred)
print(mse)
```

```python
# rmse (root mean square error ) -. square root of mse
# rmse=root_mean_squared_error(y_true,y_pred)
rmse=np.sqrt(mse)
print(rmse)
```

```python
# r2 score 
r2=r2_score(y_true,y_pred)
print(r2)
```

```python
# lasso vs ridge regression 
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso
from sklearn.metrics import mean_absolute_error

data={
    "Size":[1000,1200,1500,1800,2000,2200,2500,2700,3000,3200],
    "bedrooms":[2,2,3,3,3,4,4,4,5,5],
    "age":[20,18,15,12,10,8,6,5,4,2],
    "price":[30,35,45,50,55,60,68,72,80,85]
}
df=pd.DataFrame(data)
df
```

```python
# spiltting of data in x,y 
x=df.drop('price',axis=1)
y=df['price']

# spilt data in training and testing 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

# training --> ridge regression 
ridge=Ridge(alpha=1.0)

ridge.fit(x_train,y_train)

# prediction 
ridge_p=ridge.predict(x_test)
print("x test")
print(x_test)
print("prediction")
print(ridge_p)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mae=mean_absolute_error(y_test,ridge_p)
print("mae= ",mae)
mse=mean_squared_error(y_test,ridge_p)
print("mse",mse)

rmse=np.sqrt(mse)
print("rmse",rmse)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,ridge_p)
print("r2 score",r2)
```

```python
# lasso 
# spiltting of data in x,y 
x=df.drop('price',axis=1)
y=df['price']

# spilt data in training and testing 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

# training --> ridge regression 
lasso=Lasso(alpha=1.0)

lasso.fit(x_train,y_train)

# prediction 
lasso_p=lasso.predict(x_test)
print("x test")
print(x_test)
print("prediction")
print(lasso_p)
from sklearn.metrics import mean_absolute_error,mean_squared_error
mae=mean_absolute_error(y_test,lasso_p)
print("mae= ",mae)
mse=mean_squared_error(y_test,lasso_p)
print("mse",mse)

rmse=np.sqrt(mse)
print("rmse",rmse)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,lasso_p)
print("r2 score",r2)
```

```python
# logistic regression --> NO algorithm only classification
# when output has 2 classes --> either pass or fail
# when output has 2 answers --> either yes or no 
import pandas as pd
from sklearn.linear_model import LogisticRegression

data={
    "hours":[1,2,3,4,5,6,7],
    "result":[0,0,0,1,1,1,1]
}
df=pd.DataFrame(data)
df

# split 
x=df[['hours']]
y=df['result']

#  model 
model=LogisticRegression()
model.fit(x,y)

# prediction 
new_Data=pd.DataFrame({
    "hours":[3]
})
pred_d=model.predict(new_Data)
print(pred_d[0])

# probability prediction  
probData=model.predict_proba(new_Data)
print(probData)
```

```python
# read about classification 
# KNN ----> 
import pandas as pd 
from sklearn.neighbors import KNeighborsClassifier

# data set 
data={
    "weight":[100,110,120,170,180,190],
    "fruits":[0,0,0,1,1,1]  # 0--> apple,1--> orange
}

df=pd.DataFrame(data)
# split X and y 
x=df[['weight']]
y=df['fruits']
# model 
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x,y)

# prediction 
new_DATA=pd.DataFrame({
    "weight":[16]
})
pred_data=model.predict(new_DATA)
pred_data[0]

if(pred_data[0]==1):
    print("Orange")
else:
    print("Apple")
```
