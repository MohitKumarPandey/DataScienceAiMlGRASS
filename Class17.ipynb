```python
import numpy as np 
import pandas as pd 
url="https://raw.githubusercontent.com/05mohitkumar146-lgtm/DataScienceGrass/main/student-data.json"
df=pd.read_json(url)
df
```

```python
# by default soting is asending
# for desencding we need to write asending=False
# delete ----> 1. Hard Delete 2.Soft Delete-+
df.sort_values('english')
# desending
df.sort_values('english',ascending=False) # not on original data 

# df.sort_values('name')

# by default ascending method 1 
df.sort_values(by=['english'],ascending=[False])

# multiple columns sorting
df.sort_values(by=['english','maths'],ascending=[False,True])
df['name']=df['name'].str.lower()
df.sort_values('name')
```

```python
# add,update,delete
# add Column total = py+math+english 
df['Total']=df['english']+df['maths']+df['physics']
df
# add row 
df.loc[14]=['monu','Male',92,93,89,274]
df
# update column 

# update row

# delete column 
df.drop("Total",axis=1)
# delete row 
df.drop(14,axis=0)
df

# delete row and column 
```

```python
# Missing Values
# df.drop([6,7,8,9,10,11,12,13,14],inplace=True)
df["dateOfJoin"]=['2025-01-01','2025-02-02','2025-03-03','2025-04-04','2025-05-05','2025-06-06']
df
# string to date format 
df['dateOfJoin']=pd.to_datetime(df['dateOfJoin'])
df['dateOfJoin'].dtype
```
