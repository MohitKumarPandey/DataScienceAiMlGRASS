```python
# list
import numpy as np
l=[1,2,3]
print(l)

# array 1d
arr=np.array(l)
print(arr)
```

```python
# 2d list
l=[[1,2,3],[4,5,6]]
print(l)
print(l[0]) # [1,2,3] -->1 row
print(l[0][0])  #  [1,2,3] -->1 row --> 1 --> 1st element

# 2d array
# arr=np.array(l)
# arr=np.array([[1,2,3],[4,5,6]])
arr=np.arange(1,7).reshape(2,3)
print(arr)
print(arr[0]) # [1,2,3] -->1 row
print(arr[0][0])  #  [1,2,3] -->1 row --> 1 --> 1st element
# arr[row][column]
l[1][0]=100
arr[1][0]=100
print(l)
print(arr)
```
