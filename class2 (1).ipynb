```python
# operators 
# ternary operator does not exist in python
# 1. Arithmetic operator
# +,-,*,/,%,**,// 
a=10
b=20
print(a+b)
# -
c=20
d=10
print(c-d)
# *
e=10
f=21
print(e*f)
# /
num_1=20
num_2=2
print(num_1/num_2)
# // -> float division
num_3=20
num_4=2
print(num_3//num_4)
# %
num_5=45
num_6=9
print(num_5%num_6)
# **
num_7=45
num_8=2
print(num_7**num_8)
```

```python
# relational operator
# answer is always either true or false
# ==,<,>,>=,<=,!=

# ==
num1=10
num2=20
print(num1==num2)
# < 
num1=10
num2=11
print(num1<num2)
# >
num1=20
num2=10
print(num1>num2)
# <=
 
num1=10
num2=11
print(num1<=num2)
# >= greater than a
num1=20
num2=10
print(num1>=num2)
# != not equal to 
num1=12
num2=13
print(num1!=num2)
```

```python
# assignment operator
# =,+=,-=,*=,/=,//=,%=
# = -> use into declare a variable
a=10
b=a+20
print(b)
a+=20
print(a)
# -=
c=30
print(c)
c-=10
print(c)
# *=
d=10
print(d)
d*=2
print(d)
# //=
e=12
e//=5
print(e)
# /=
f=34
f/=2
print(f)
```

```python
# logical operator
# and ,or,not

# number ---> +ve and -ve -> true  and 0 -->false
# and
a=10
b=20
print(a and b)  # if both value is true then 2nd value is return
# if first or second value is false then first or second value is print
# or
print(a or b)  # if both values are true then first value is return otherwise which value is true then that true value is print

# not
c=0
print(not(c))


name1="hello"
name2=""
print(name1 and name2)
print(name1 or name2)
print(not(name1))
```
