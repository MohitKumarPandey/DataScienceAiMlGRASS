```python
# polynomial regression 
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# data 
df=pd.DataFrame({
    "Experience":[1,2,3,4,5]
})
# x=df[['Experience']]
x=df
y=[10,20,50,90,150]

# Poly features 

poly=PolynomialFeatures(degree=3)
x_poly=poly.fit_transform(x)
print(x_poly)
# model 
model=LinearRegression()
model.fit(x_poly,y)

# prediction 
input_data=int(input("enter the experience"))
new_data=pd.DataFrame({
    "Experience":[input_data]
})
# convert new data into poly 
new_poly=poly.transform(new_data)
predicted_data=model.predict(new_poly)
print(predicted_data[0])
```

```python
plt.plot(x_poly,y)
```

```python
# under and Overfitting 
# LinearRegression---> under fitting 
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression
df=pd.DataFrame({
    "linear":[1,2,3,4,5]
})
x=df 
y=[10,20,50,90,150]

# model 
model=LinearRegression()
model.fit(x,y)

# model prediction 
input_data1=int(input("enter the label:-"))
new_data1=pd.DataFrame({
    "linear":[input_data1]
})

predict_data=model.predict(new_data1)
print(predict_data[0])
```

```python
# OverFitting
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
df=pd.DataFrame({
    "label":[1,2,3,4,5]
})
x=df 
y=[10,20,50,90,150]
# poly convert 
poly=PolynomialFeatures(degree=3)
x_poly=poly.fit_transform(x)

# model 
model=LinearRegression()
model.fit(x_poly,y)
# model prediction 
input_data2=int(input("enter the label:-"))

new_data2=pd.DataFrame({
    "label":[input_data2]
})
# input data to polynomial 
u_new_DATA=poly.transform(new_data2)

# prediction 
predicted_data1=model.predict(u_new_DATA)
print(predicted_data1[0])
```

```python
plt.plot(x_poly,y)
```

```python
# classification example 
import numpy as np 
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
# data
df=pd.DataFrame({
    "Weight":[120,140,150,170],
    "Size":[6,7,7,8]
})
x=df 
y=['Apple','Apple','Orange','Orange']

# model 
model=DecisionTreeClassifier()
model.fit(x,y)

# prediction 
input_weight=int(input("enter the weight"))
input_size=int(input("enter the size"))

new_Data=pd.DataFrame({
    "Weight":[input_weight],
    "Size":[input_size]
})

# prediction 
predict_Data=model.predict(new_Data)
print(predict_Data[0])
```

```python
# pass 40 above --> pass | less than 40 fail 
data=pd.DataFrame({
    "marks":[10,20,25,39,40,55,56,70,80,91],
    "result":['fail','fail','fail','fail','pass','pass','pass','pass','pass','pass']
})
x=data[['marks']]
y=data['result']

# model dec 
model=DecisionTreeClassifier()
model.fit(x,y)
# prediction data
input_marks=int(input("enter the marks"))
new_MarksData=pd.DataFrame({
    "marks":[input_marks]
})

predict_marks=model.predict(new_MarksData)
print(predict_marks[0])
```
