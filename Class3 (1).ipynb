```python
# list
num1 = [] # default [] -> false || [1,2,3] -> true
print(not(num1))


# dictionary
num2={}   # default -> false || {1,2,3} -> true
print(not(num2))
```

```python
# [] -> list is define  by  it
# empty list -> false
# [1,2] ->true
# {} -> dictionary is define by it
# empty dictionary -> false
# {1} -> true
```

```python
# identity operators
# is | is not
name="hello"
name1="hello"
print(name==name1)

print(name is name1) # value | address
print(id(name))
print(id(name1))

# example 2  (-5 to 256 )-> then same memory is point -> applicable for is operator
# if value is reach out of this range then (is) operator return false because both address is changed
num1=-6
num2 =-6
print(num1==num2)
print(num1 is num2)
print(id(num1))
print(id(num2))

# example 3
num1=10
num2=10.0
print(num1==num2)
print(num1 is num2)
example 4
num1=[1,2,3]
num2=[1,2,3]
num3=num1
print(num1==num2)
print(num1 is num2)
print(id(num1))
print(id(num2))
print(id(num3))
```

```python
# conditonal statements
# if 
# elif
# else


# if
num1=10
num2=20
if num1<num2:
    print(num2)

# if-else
num1=int(input("enter the number"))
if num1:
    print(num1)
else:
    print("zero",num1)

# if elif else
age=int(input("enter the age"))
if age==18:
    print("your age is" ,age)
elif age>18:
    print("you are above 18: ",age)
else:
    print("you are less than 18 : ",age)

# ques on if elif and else
num1=int(input("enter the number"))
if num1>0:
    print(num1,"is positive")
elif num1<0:
    print(num1,"is negative")
else:
    print("number is zero")

```
