```python
# example 4
class Student:
    def __init__(self,*args):
        print(type(args))
        print(args)
        self.name=args[0]
    def getStu(self):
        # print("the student is ",self.name)
        return self.name
s=Student("Mohit",20,"00000000","mohit@gmail.com")
t=s.getStu()
print(t)
```

```python
# example 6
class Student:
    def __init__(self,*args):
         
        self.data=args
    def users(self):
        for i in self.data[0]:
            print(i)
    def details(self):
        for i in self.data[1]:
            # print(self.data[1][i])
            print(f"{i}= {self.data[1][i]}")
s=Student(["Dheeraj","kunal","harsh","praveen"],{"address":"kukas","college":"arya","loc":"jaipur"})
s.users()
s.details()
```

```python
class Bank:
    def __init__(self,args=0):
        self.amount=args
    def deposit(self,amount):
        if(amount<=0):
            print("invalid amount")
        else:
            self.amount+=amount
    def withdraw(self,amount):
        if(amount<=0 or amount>self.amount):
            print("insufficient balance")
        else:
            self.amount-=amount
    def showbalance(self):
        return self.amount
b=Bank(1000)
b.deposit(200)
print(b.showbalance())
b.withdraw(100)
print(b.showbalance())
# b.deposit(-1)
print(b)
d=Bank()
d=b
print(d.showbalance())
```
