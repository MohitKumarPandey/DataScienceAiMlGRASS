```python
# Import modules
# function is  work for everyone
# methods are specific to person 

# modules  
# import math  -> math is libraries/modules/package
# from math import pi -> then we not need to write math.pi() direct use pi()
# from math import * -> use everything from math librarie without need to write math.
# import math as raj -> here raj is the nickname of math . now we can access all methods using raj



# example
import math
print(math.factorial(5))
print(math.pow(3,3))
print(math.pi)
print(math.floor(10.7324))
print(math.ceil(10.13))
print("\n")
# example 2
import math as mohit
print(mohit.factorial(5))
print(mohit.pow(3,3))
print(mohit.pi)
print(mohit.floor(10.7324))
print(mohit.ceil(10.13))

# example 2
print("\n")
from math import factorial,pi
print(factorial(6))
print(pi)
# example 3
from math import *
print(pow(5,4))
print(factorial(3))
print(floor(10.234))
print(ceil(12.244))
print(pi)
```
