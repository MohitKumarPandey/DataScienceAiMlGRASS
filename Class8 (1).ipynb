```python
# encapsulation
class Address:
    def __init__(self,city,state):
        self.__city=city
        self.state=state
    def show(self):
        print(f"my location is {self.__city} in {self.state}")
    def getCity(self):
        return self.__city
a=Address("chhapra","Bihar")
a.show()
# print(a._Address__city) # _  --> use for protected and __ --> use for private
print(a.getCity())
print(a.state)
```
