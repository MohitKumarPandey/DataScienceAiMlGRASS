```python
# Project 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
url="https://raw.githubusercontent.com/rajendra0968jangid/Ds-arya/main/temperature_data.json"
df=pd.read_json(url)
df
```

```python
# task 1 cleaning of data
# drop row 3 --> day=wednesday
df.dropna(subset='temperature_c',inplace=True)
df
# fill average value --> humidity --> nan
# df.loc[6,'humidity_pct']=df['humidity_pct'].mean()
# df

df['humidity_pct']=df['humidity_pct'].fillna(df['humidity_pct'].mean())
# new cols --> farenheit 
df['temperature_F']=(df['temperature_c']*9)/5+32
df
```

```python
# subplots --> pie 
fig,aux=plt.subplots(2,2)
aux[0,0].plot(df['humidity_pct'],df['temperature_c'])
aux[0,0].set_xlabel("Humidity")
aux[0,0].set_ylabel("Temperature in celcius")
aux[0,0].set_title("Humidity with celcius")

aux[0,1].pie(df['temperature_F'],labels=df['day'],autopct='%1.2f%%')
aux[0,1].set_title("Fahrenheit chart")

aux[1,0].pie(df['humidity_pct'],labels=df['day'],autopct='%1.1f%%')
aux[1,0].set_title("humidity with day")


aux[1,1].plot(df['temperature_c'],df['temperature_F'],marker='o')
aux[1,1].set_ylabel("Fahrenhiet")
aux[1,1].set_xlabel("Celcius")
aux[1,1].set_title("Celcius vs Fahrenheit")
aux[1,1].grid(True)
plt.tight_layout()
# save image 
plt.savefig("Analysis.png")
```
