```python
# Multiple line chart
x=[2010,2015,2020,2025]
y1=[100,200,260,290]
y2=[150,185,195,300]
plt.plot(x,y1,label='jeans')
plt.plot(x,y2,label='shirt')
plt.legend() # it is use to describe the line meaning
plt.show()
```

```python
# ques 
# multiple line chart 
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
product_A = [120, 135, 150, 170, 180, 200]
product_B = [100, 110, 130, 125, 145, 160]
product_C = [80, 95, 100, 120, 140, 155]
plt.figure(figsize=(6,4),dpi=90)
plt.plot(months,product_A,label='Product A',marker='o')
plt.plot(months,product_B,label='Product B',marker='s')
plt.plot(months,product_C,label='Product C',marker='*')
plt.title("Monthly product sales ")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
```

```python
# bar chart 
import matplotlib.pyplot as plt 
x=[2010,2015,2020,2025]
y=[100,150,200,290]
plt.figure(figsize=(6,4),dpi=80)
plt.bar(x,y)
```

```python
# multi bar chart
x=[1,2,3,4]
arr=np.array(x)
y1=[10,20,20,40]
y2=[20,30,25,30]
# calculations 
w=0.40
plt.figure(figsize=(6,4),dpi=90)
plt.bar(arr-w/2,y1,label='Boys',width=w)
plt.bar(arr+w/2,y2,label='Girls',width=w)
plt.xlabel("Groups")
plt.ylabel("No. of students")
plt.title("No of group in each group")
plt.legend()
```
