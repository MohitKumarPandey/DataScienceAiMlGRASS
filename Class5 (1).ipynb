```python
# we need to pass 2 arguement in map(function,list),reduce(),filter()
#map(transform) ->  the exixting list is remain change and new list is created  . It is use to perform similar operation
def square(x):
    return x*x
l=[10,20,30]
l1=list(map(square,l)) # an internal loop chalega len(list) tk
print(l1)
# example 
l=[10,20,30]
l1=list(map(lambda x:x*x,l))
print(l1)

# alternative method
l1=[]
for i in range(len(l)):
    l1.append(l[i]*l[i])
print(l1)
# filter(select values)  
def hello(x):
    return x.startswith('a')
l=["apple","banana","cat","dog"]
l1=list(filter(hello,l))
print(l1)
# example
l2=[]
for i in l:
    if 'a'==i[-1]:
        l2.append(i)
print(l2)

# convert the list elements in upper case
 
l3=list(map(lambda x:x.upper(),l))
print(l3)
# reduce(combine value)
```
