```python
#FUNCTION
# def keyword is use to define a function
# 1. arguement -> written inside the () at the time of function call . Pass value
# 2. parameter -> written inside the () at the time of function definition. Receive value 

# it must be declare the fuction before calling
def hello():
    print("hello function is working")
hello()

# example 2
def hello(a):   # a is parameter
    print(a)
hello(100)

# example 
# use * in case when multiple arguement is pass
def student(*a):
    # if we want to access 'a' outside the function then we need to make it global
    print(a)
    print(type(a))
student(1,2,3,4,5,6)

# in python if 2 functions are have same name then the last declare function is call 
```

```python
# lambda function [lambda arguement:expression]
add =lambda x:x
print(add(100))

sum = lambda x,y:x+y
print(sum(10,20))

a= lambda x:x
print(a([10,20,30,40,50]))

star=lambda *x:x
print(star(10,20,30,40,50))
```

```python
# list comprehension [expression for item in iterable]
print([i for i in range(5)])

# example 2
l=[10,20,30,40,50]
print(l)
print([l[i] for i in range(len(l))])


# example 3
l=[10,20,[30,40,50,60]]
print([l[2][i] for i in range(len(l[2]))])
```

```python
# list
# len(), .append(el), .insert(index,el)

l=[10,20,45,89,78,"hello",True]
print(len(l))
print(l)
# append()
l.append(90)
print(l)
# .insert()
l.insert(3,"python")
print(l)
```
