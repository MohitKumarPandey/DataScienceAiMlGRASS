```python
# data splitting in training and testing
import numpy as np 
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
data={
  'experience':[300,600,900,1500],
    'salary':[1000,1500,2000,3000]
}
df=pd.DataFrame(data)
df
```

```python
# split data
X=df['experience']
y=df['salary']
# train test split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
print("x train",x_train)
print("x test",x_test)
print("y train",y_train)
print("y test",y_test)
```

```python
url='https://raw.githubusercontent.com/rajendra0968jangid/Ds-Arya/main/extended_salary_data.csv'
Newdata=pd.read_csv(url)
Newdata
X=Newdata[['experience']]
y=Newdata['salary']
# train test split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=3)
print("x train",x_train)
print("x test",x_test)
print("y train",y_train)
print("y test",y_test)
```

```python
# simple linear regression 
from sklearn.linear_model import LinearRegression

# model fit
model=LinearRegression()
model.fit(x_train,y_train)
user=int(input("enter your experience"))
# model prediction 
new_data={
    'experience':[user]
}
df1=pd.DataFrame(new_data)

# model prediction 
predict_data=model.predict(df1)
print(predict_data[0])
```
